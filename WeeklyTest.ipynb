```python
def count_wrong(text):
    text = text.strip()

    if text == "":
        return 0

    if text.isdigit():
        return int(text)

    return len(text.split())


def input_wrong_numbers(text, valid_numbers):
    text = text.strip()

    if text == "":
        return []

    nums = list(map(int, text.split()))
    return [n for n in nums if n in valid_numbers]


def make_reading_book_sentence(n):
    if n == 0:
        return ""
    return f"이번 주에 읽어야 하는 책의 내용을 활용한 {n}문제를 틀렸습니다."


def make_reading_nonfiction_sentence(wrong_nums):
    reading = [n for n in wrong_nums if 4 <= n <= 6]
    nonfiction = [n for n in wrong_nums if 7 <= n <= 9]

    total = len(reading) + len(nonfiction)

    if total == 0:
        return ""

    if reading and nonfiction:
        area = "독서, 비문학"
    elif reading:
        area = "독서"
    else:
        area = "비문학"

    return f"주어진 지문 안에서 근거를 찾는 문제({area}) 부분에서 {total}개를 틀렸습니다."


def make_detail_sentence(area_name, wrong_nums, descriptions):
    if not wrong_nums:
        return ""

    parts = []

    for num in wrong_nums:
        desc = descriptions.get(num)
        if desc:
            parts.append(f"{desc} 1개")

    if not parts:
        return ""

    body = ", ".join(parts)

    return f"{area_name}은 {body}를 틀렸습니다."


def make_vocab_sentence(wrong_19, wrong_20_count):
    parts = []

    if wrong_19:
        parts.append("사자성어의 뜻을 찾는 1문제")

    if wrong_20_count > 0:
        parts.append(f"단어의 뜻을 찾는 {wrong_20_count}문제")

    if not parts:
        return ""

    body = ", ".join(parts)

    return f"어휘는 {body}를 틀렸습니다."
```

```python
descriptions = {
    10: "소설의 구성방식 중 '삽화'의 개념을 묻는 문제",
    11: "소설의 구성방식 중 '의식의 흐름 구성'을 판단하는 문제",
    12: "시의 화자를 판단하는 문제",
    13: "고전의 화자를 판단하는 문제",
    14: "시의 표현법을 묻는 문제",

    15: "'접미사'를 통해 품사가 바뀐 단어를 찾는 문제",
    16: "문장 성분을 판단하는 문제",
    17: "품사를 구분하는 문제",
    18: "문법 개념을 적용하는 문제"
}
```

```python
# 입력하지 않은 부분은 모두 정답 처리됨

book_wrong_count = count_wrong(
    input("1~3번 틀린 개수 입력 / 없으면 Enter: ")
)

reading_wrong_nums = input_wrong_numbers(
    input("4~9번 중 틀린 번호 입력 예: 4 7 9 / 없으면 Enter: "),
    range(4, 10)
)

literature_wrong_nums = input_wrong_numbers(
    input("10~14번 문학 중 틀린 번호 입력 예: 10 11 / 없으면 Enter: "),
    range(10, 15)
)

grammar_wrong_nums = input_wrong_numbers(
    input("15~18번 문법 중 틀린 번호 입력 예: 15 18 / 없으면 Enter: "),
    range(15, 19)
)

wrong_19 = input("19번 사자성어 문제를 틀렸으면 x 입력 / 맞으면 Enter: ").strip().lower() == "x"

wrong_20_count = count_wrong(
    input("20번 단어 뜻 문제 틀린 개수 입력 / 없으면 Enter: ")
)
```

```python
literature_descriptions = {
    num: descriptions[num]
    for num in range(10, 15)
}

grammar_descriptions = {
    num: descriptions[num]
    for num in range(15, 19)
}

sentences = [
    make_reading_book_sentence(book_wrong_count),
    make_reading_nonfiction_sentence(reading_wrong_nums),
    make_detail_sentence("문학", literature_wrong_nums, literature_descriptions),
    make_detail_sentence("문법", grammar_wrong_nums, grammar_descriptions),
    make_vocab_sentence(wrong_19, wrong_20_count)
]

final_text = " ".join([sentence for sentence in sentences if sentence])

print(final_text)
```
